```python
import matplotlib.pyplot as plt
import numpy as np

# Define Amdahl's Law function
def amdahls_law(P, N):
    return 1 / ((1 - P) + (P / N))

# Vector speedup
N = 10

# Percentage of vectorization from 0 to 100%
P_values = np.linspace(0, 1, 100)

# Calculate speedup values
speedup_values = amdahls_law(P_values, N)

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(P_values * 100, speedup_values, label='Amdahl\'s Law')
plt.xlabel('Percent Vectorization')
plt.ylabel('Net Speedup')
plt.title('Net Speedup vs Percent Vectorization')
plt.grid(True)
plt.legend()
plt.show()
```
